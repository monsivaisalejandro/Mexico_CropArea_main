```python
#read Sentinel1 images
os.chdir(Sentinel1_image_folder)
SAR_images = glob.glob("*.zip")
SAR_images = sorted(SAR_images)

print('Number of images:',len(SAR_images)) #list the existing Sentinel-1 images
#print(SAR_images)
```

```python
proj = '''GEOGCS["WGS 84",DATUM["WGS84(DD)",SPHEROID["WGS84",6378137.0,298.257223563,\
AUTHORITY["EPSG","7030"]],AUTHORITY["EPSG","6326"]],PRIMEM["Greenwich",0,AUTHORITY["EPSG","8901"]],\
UNIT["degree",0.0174532925199433,AUTHORITY["EPSG","9122"]],AXIS["Latitude",NORTH],AXIS["Longitude",EAST],AUTHORITY["EPSG","4326"]]'''
```

```python
for images in SAR_images:
    zip_file_name=images
    path_input_file=os.path.join(Sentinel1_image_folder,zip_file_name)
    #print(path_input_file)

    # 1. Get the Java File type
    File = jpy.get_type('java.io.File')
    #2. Read zip file containing Sentinel-1 images
    java_file = File(str(path_input_file))
    # 3. Read the product
    product = ProductIO.readProduct(java_file)
    
    if not product:
        print("Failed to read product.")
    else:
        print(f"Reading product Name: {product.getName()}")
    
    # Use HashMap for SNAP parameters
    HashMap = jpy.get_type('java.util.HashMap')
    parameters = HashMap()

    #Sentinel-1 image processing
    applyorbit = s1f.do_apply_orbit_file(product)
    thermaremoved = s1f.do_thermal_noise_removal(applyorbit)
    pols = 'VH,VV' #Sentinel-1 polarizations
    polarization = 'DV' # Dual-polarization V
    calibrated = s1f.do_calibration(thermaremoved, polarization, pols)
    down_filtered = s1f.do_speckle_filtering(calibrated)
    down_tercorrected = s1f.do_terrain_correction(down_filtered, proj, 1)
    down_subset = s1f.do_subset(down_tercorrected, wkt)
    del applyorbit
    del thermaremoved
    del calibrated
    del down_filtered
    del down_tercorrected
    
    output_file=os.path.join(output_dir_TIFF,down_subset.getName())
    print("Writing ",output_file)
    #1. Write BEAM-DIMAP file
    ProductIO.writeProduct(down_subset, output_file, 'BEAM-DIMAP')
    #2. Write TIFF file
    ProductIO.writeProduct(down_subset, output_file, 'GeoTIFF')
    #ProductIO.writeProduct(down_subset, output_file, 'GeoTIFF-BigTIFF')
    product.dispose()
    product.closeIO()

    down_subset.dispose()
    del down_subset
```

```python
def zip_directory(directory_path, zip_name):
    """
    Creates a compressed ZIP file from a local directory, maintaining relative paths.
    """
    # Use a context manager to ensure the zip file is closed properly
    with zipfile.ZipFile(zip_name, 'w', zipfile.ZIP_DEFLATED) as zipf:
        # Walk through the directory tree
        for root, dirs, files in os.walk(directory_path):
            for file in files:
                file_path = os.path.join(root, file)
                # Calculate the relative path so the archive structure mirrors the folder
                archive_file_path = os.path.relpath(file_path, start=directory_path)
                zipf.write(file_path, archive_file_path)
    
```
